```python
import tellurium as te
```

```python
model = '''
model lacMetabolism
    # Declare species
    species lac;
    
    # Break down of lactose into glucose and galactose by bGal
    lac -> glu + gal; (Vm_bGal * lac) / (km_bGal + lac)

    # Species Initialization
    lac = 0; gal = 0;
    # Parameters
    Vm_bGal = 1; km_bGal = 1;
end
'''
```

```python
test = model + '''
model test
    R: lacMetabolism()

    R.lac is lac
    # Injection of lactose at t=5
    at (time >5): lac = 5
end
'''


rr = te.loadAntimonyModel(test)
result = rr.simulate(0, 25, 100)
p = te.plotWithLegend(rr, result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def makeNoise(results, mean=0, std=1):
    import numpy as np
    dim = results.shape
    numSpecies = dim[1] - 1
    numPoints = dim[0]
    noise = np.hstack([np.array([np.random.normal(mean, std, numPoints)]).T for s in range(numSpecies)])
    return noise

noise = makeNoise(result, std=0.2)
observed = np.hstack((np.array([result[:, 0]]).T, result[:, 1:] + noise))

plts = plt.plot(result[:, 0], result[:, 1:], linewidth=2.5)
for i, column in enumerate(observed[:, 1:].T):
     plt.scatter(
                 observed[:, 0],
                 column,
                 linewidth=2.5,
                 color=plts[i].get_color())
```
